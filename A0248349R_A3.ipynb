```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("repository_data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape[0]
```

```python
df.describe()
```

```python
df_sorted = df.sort_values(by="stars_count", ascending=False)
df_sorted.head()
```

```python
plt.figure(figsize=(60,20))
sns.barplot(x=df_sorted["stars_count"][:15], y=df_sorted["name"][:15])

plt.title("\n Top 15 stared github repositories \n", size=36, color="black")
plt.xticks(fontsize=32)
plt.xticks(fontsize=24)
plt.xlabel("\n Stars Count \n", fontsize=32, color="black")
plt.ylabel("\n repository name \n", fontsize=32, color="black")

plt.savefig("repositories ordered by stars_count")
plt.show()
```

```python
df_sorted = df.sort_values(by="forks_count", ascending=False)
df_sorted.head()
```

```python
plt.figure(figsize=(60,20))
sns.barplot(x=df_sorted["forks_count"][:15], y=df_sorted["name"][:15])

plt.title("\n Top 15 forked github repositories \n", size=36, color="black")
plt.xticks(fontsize=32)
plt.xticks(fontsize=24)
plt.xlabel("\n Forks Count \n", fontsize=32, color="black")
plt.ylabel("\n repository name \n", fontsize=32, color="black")

plt.savefig("repositories ordered by forks_count")
plt.show()
```

```python
pie_label = [language for language in df["primary_language"].value_counts().sort_values(ascending=False).index]
```

```python
plt.figure(figsize=(60,20))
plt.pie(df["primary_language"].value_counts().sort_values(ascending=False)[:10], labels=pie_label[:10], autopct='%1.2f%%', textprops={'fontsize': 20})
plt.legend(loc='upper left',bbox_to_anchor=(-0.1,1), fontsize=22)
plt.title("\n The top 10 most popular primary languages \n", color="black", size=36)
plt.savefig("The Top 10 most popular primary languages")
plt.show()
```

```python
df_corr = df.drop(["name", "primary_language", "languages_used", "created_at", "licence"], axis=1)
df_corr.head()
corr_matrix = df_corr.corr()
```

```python
plt.figure(figsize=(60,20))
plt.rc("font",family="Times New Roman",size=24)
sns.set(font_scale = 1.5)
sns.heatmap(corr_matrix, vmin=-1, vmax=1, annot=True)
plt.title("\n The correlation between these counts \n", color="black", size=36)
plt.savefig("Correlation")
plt.show()
```
